<h1>Import Library yang digunakan</h2>

```
import os
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

<h1>Mengabungkan Director dengan path File</h2>

```
director = []
for i in os.listdir():
  direc_dir = os.path.join(os.getcwd(),i)
  director.append(direc_dir)
```

<h1>Menampilkan file yang ada di director</h2>

```
print(director)
```

<h1>Membaca file dari index ke 3 "train.csv" dan menampilkan isinya</h2>

```
data_train = pd.read_csv(director[3])
print(data_train.head(10))
```

<h1><b>1. Missing Value pada LotFrontage </h1>

```
TOTAL_NULLDATA = data_train['LotFrontage'].isnull().sum()
print(f"TOTAL DATA YANG HILANG PADA BARIS LotFrontage  : {TOTAL_NULLDATA} DATA")
```

<h1>MEMBUAT SETIAP NILAI Nan menjadi terisi dengan nilai median dengan round angka dibelakang mejadi 1 digit saja</h1>

```
nulll_Value_to_Median = data_train['LotFrontage'].fillna(data_train['LotFrontage'].mean().round(1),inplace=True)
```

<h1>Menampilkan hasil setelah melakukan Missing Values</h2>

```
TOTAL_NULLDATA = data_train['LotFrontage'].isnull().sum()
print(f"TOTAL DATA YANG HILANG PADA BARIS LotFrontage  : {TOTAL_NULLDATA}")
print(data_train['LotFrontage'].head(20))
```

<h1><b>2. ENCODING Pada Kolom NeighborHoot<b></h1>

```
data_train.head(20)
```

```
dataset_NeighborHoot = LabelEncoder()

# Transformasi kolom 'Neighborhood'
data_train['Neighborhood'] = dataset_NeighborHoot.fit_transform(data_train['Neighborhood'])

print(data_train['Neighborhood'].head(10))
```
